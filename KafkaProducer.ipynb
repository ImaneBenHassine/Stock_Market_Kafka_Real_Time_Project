```python
import pandas as pd
from kafka import  KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
producer = KafkaProducer(bootstrap_servers =['13.48.133.6:9092'],
                         value_serializer = lambda x :
                         dumps(x).encode('utf-8'))

```

```python
producer.send('demo_test', value ="{'hello :kafka :world: ibh'}")
```

```python
df = pd.read_csv ("indexProcessed.csv")
```

```python
# convert data to json format

while True:
  dict_stock = df.sample(1).to_dict(orient="records")[0]
  producer.send('demo_test', value =dict_stock)
  sleep(1) #add some delay so we don't overload our server so we will add delay of one second you
```

```python
producer.flush() #clear data from kafka server
```
